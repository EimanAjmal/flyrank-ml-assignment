## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*
**Decision this improves:** which pages a content team spends their limited
review time on first, out of potentially hundreds of pages.

**Unit of analysis:** one content page (content_id), evaluated using its
90-day performance window.

**Action someone takes:** a content editor opens the top-ranked pages and
decides whether to refresh, expand, or leave the page as-is.

**Cost of a wrong recommendation:**
- False positive (page flagged as needing refresh, but it's actually fine):
  wastes reviewer time on a page that didn't need attention.
- False negative (page truly declining, but not flagged): a real opportunity
  to recover traffic gets missed, and the page keeps losing visibility unnoticed.

**Why data/ML can help:** manually reviewing hundreds of pages to spot patterns
like "declining despite demand" is slow and inconsistent. A ranked,
evidence-based score lets a small team focus first on pages most likely to
be worth the effort.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

url = "https://raw.githubusercontent.com/EimanAjmal/flyrank-ml-assignment/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# Number 1: how many pages qualify (impressions_90d > 0, content_age_days >= 90)
qualifying = df[(df['impressions_90d'] > 0) & (df['content_age_days'] >= 90)]
print(f"Qualifying pages: {len(qualifying)} out of {len(df)} total")

# Number 2: how many are currently declining
declining_pct = (qualifying['trend_direction'] == 'down').mean() * 100
print(f"% of qualifying pages currently declining: {declining_pct:.1f}%")

# Number 3: stale-but-visible pages (a Lane 2 reason code)
stale_visible = qualifying[(qualifying['days_since_last_update'] >= 180) &
                             (qualifying['impressions_90d'] >= 500)]
print(f"Stale + visible pages (candidates for refresh): {len(stale_visible)}")

```
